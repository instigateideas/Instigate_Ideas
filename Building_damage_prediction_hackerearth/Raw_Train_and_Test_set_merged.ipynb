```python
for i in build_struct.columns:
    if i in train_data.columns:
        print(i)
```

```python
build_struct=build_struct.drop(labels=['district_id','vdcmun_id'], axis=1)
```

```python
build_owner.columns
```

```python
for i in build_owner.columns:
    if i in train_data.columns:
        print(i)
```

```python
build_owner=build_owner.drop(labels=['district_id','vdcmun_id'], axis=1)
```

```python
train_data_1=train_data.merge(build_struct, on='building_id', how='left')
```

```python
train_data_1.head()
```

```python
train_data_1.columns
```

```python
for i in build_owner.columns:
    if i in train_data_1.columns:
        print(i)
```

```python
train_data_1=train_data_1.drop(labels='ward_id', axis=1)
```

```python
train_data_2=train_data_1.merge(build_owner, on='building_id',how='left')
```

```python
train_data_2.columns
```

```python
train_data_2.head()
```

### Test data merging

```python
for i in build_owner.columns:
    if i in test_data.columns:
        print(i)
```

```python
test_data_1=test_data.merge(build_owner, on='building_id', how='left')
test_data_1.shape
```

```python
for i in build_struct.columns:
    if i in test_data_1.columns:
        print(i)
```

```python
test_data_1=test_data_1.drop('ward_id', axis=1)
```

```python
test_data_2=test_data_1.merge(build_struct, on='building_id', how='left')
```

```python
test_data_2.head()
```

### Saving the joined file

```python
train_data_2.to_csv('train_data_merged.csv')
test_data_2.to_csv('test_data_merged.csv')
```
